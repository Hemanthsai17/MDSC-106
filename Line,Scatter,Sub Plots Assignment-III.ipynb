```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_mfbn=pd.read_csv('most_frequent_baby_names.csv')
df_mfbn.head()
```

```python
table=pd.pivot_table(data=df_mfbn,index="Name",values="Frequency",aggfunc=np.sum)
table
```

```python
plt.figure(figsize=(15,10))
plt.plot(table.index,table["Frequency"])
plt.xticks(rotation=90)
plt.xlabel('Name')
plt.ylabel('NUMBER OF PEOPLE WITH THAT NAME')
plt.title('Name vs NUMBER OF PEOPLE WITH THAT NAME')
plt.savefig('hwlp1.png',dpi=300)
plt.show()
```

```python
df_bs=pd.read_csv('births.csv')
df_bs.head()
```

```python
table2=pd.pivot_table(data=df_bs,index="District Name",values="Number",aggfunc=np.sum)
table2
```

```python
plt.figure(figsize=(15,10))
plt.plot(table2.index,table2["Number"])
#plt.scatter(table2.index,table2["Number"])
#plt.xticks(rotation=0)
plt.xlabel('DISTRICT NAME')
plt.ylabel('NUMBER OF BIRTHS IN THAT YEAR')
plt.title('DISTRICT NAME vs NUMBER OF BIRTHS IN THAT YEAR')
plt.savefig('hwlp2.png',dpi=300)
plt.show()
```

```python
table3=pd.pivot_table(data=df_bs,index="Neighborhood",values="Number",aggfunc=np.sum)
table3
```

```python
plt.figure(figsize=(15,10))
plt.scatter(table3.index,table3["Number"])
plt.xticks(rotation=90)
plt.xlabel('YEAR')
plt.ylabel('NUMBER OF BIRTHS IN THAT YEAR')
plt.title('YEAR vs NUMBER OF BIRTHS IN THAT YEAR')
plt.savefig('hwsp1.png',dpi=300)
plt.show()
```

```python
df_ieba=pd.read_csv('immigrants_emigrants_by_agen.csv')
df_ieba.head()
```

```python
table4=pd.pivot_table(data=df_ieba,index="District Name",values="Immigrants",aggfunc=np.sum)
table4
```

```python
table5=pd.pivot_table(data=df_ieba,index="District Name",values="Emigrants",aggfunc=np.sum)
table5
```

```python
plt.figure(figsize=(15,10))
plt.scatter(table4.index,table4["Immigrants"],color='red')
plt.scatter(table5.index,table5["Emigrants"])
#plt.xticks(rotation=90)
plt.xlabel('District Name')
plt.ylabel('Immigrants and Emmigrants')
plt.title('District Name vs Immigrants-red and Emmigrants-blue')
plt.savefig('hwsp2.png',dpi=300)
plt.show()
```

```python
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(30,20))

ax[0].scatter(table4.index,table4["Immigrants"],color='red')
ax[0].set_xlabel('District Name')
ax[0].set_ylabel('Immigrants and Emmigrants')
ax[0].set_title('District Name vs Immigrants-red and Emmigrants-blue')

ax[1].scatter(table5.index,table5["Emigrants"])
ax[1].set_xlabel('District Name')
ax[1].set_ylabel('Immigrants and Emmigrants')
ax[1].set_title('District Name vs Immigrants-red and Emmigrants-blue')

plt.show()
plt.savefig('subplotshw1.png')
```
