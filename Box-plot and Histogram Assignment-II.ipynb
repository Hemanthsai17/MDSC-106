## Box Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_birt=pd.read_csv('births.csv')
df_birt.head()
```

```python
dt={}
for i in df_birt['Year'].unique():
    dt[i]=df_birt[df_birt['Year']==i].Number
print(dt)
```

```python
plt.figure(figsize=(15,10))
plt.boxplot([x for x in dt.values()], labels=[x for x in dt.keys()])
plt.xlabel("Year")
plt.ylabel("Number of births")
plt.title("Year vs Number of births")
plt.savefig('box1hw.png')
plt.show()
```

```python
df_air=pd.read_csv('air_quality_Nov2017n.csv')
df_air.head()
```

```python
da={}
for i in df_air['Station'].unique():
    da[i]=df_air[df_air['Station']==i].NO2Value
print(da)
```

```python
plt.figure(figsize=(15,10))
plt.boxplot([x for x in da.values()], labels=[x for x in da.keys()])
plt.xticks(rotation=60)
plt.xlabel("Year")
plt.ylabel("Number of births")
plt.title("Year vs Number of births")
plt.savefig('box2hw.png')
plt.show()
```

## Histogram

```python
plt.figure(figsize=(15,10))
plt.hist(df_air['NO2Value'],rwidth=0.96,alpha=0.4,color='blue',bins=15,edgecolor='r')
plt.xlabel("NO2 Value")
plt.ylabel("Frequency")
plt.title("NO2 value frequency")
plt.savefig('hwhist1.png')
plt.show()
```

```python
df_acl=pd.read_csv('accidents_2017.csv')
df_acl.head()
```

```python
plt.figure(figsize=(15,10))
plt.hist(df_acl['Hour'],rwidth=0.96,alpha=0.4,color='blue',bins=15,edgecolor='r',histtype='stepfilled')
plt.xticks(rotation=40)
plt.xlabel("NO2 Value")
plt.ylabel("Frequency")
plt.title("NO2 value frequency")
plt.savefig('hwhist2.png')
plt.show()
```
